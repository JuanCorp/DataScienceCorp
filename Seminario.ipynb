## SCRIPT Extracción de Datos de Ecommerce

```python
import urllib
import pandas as pd
import re
import unicodedata
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
```

### 1. Extraccion de url, nombre y tipo de negocio

```python
#URL de directorio de ecommerce.
url_directorio = "http://www.ecommerce.com.do/directorio/"

page = urllib.request.urlopen(url_directorio).read()

parser = BeautifulSoup(page,'lxml')

URLs = []
Tipos = []
Nombres= []

#Encuentra todas los elementos que son tablas.
for table in parser.find_all("table"):
    #Divide el HTML por linea.
    t = str(table)
    t = t.split("\n")
    Tipo=" "
    for line in t:
        #Por cada linea en el HTML, encuentra lo que esta entre el markup h4. Eso es el Tipo de Negocio.
        if "<h4>" in line:
            search = re.search('<h4>(.*)</h4>',line )
            Tipo = search.group(1)
        #Si tiene <a y href es un url de un negocio. Extra el nombre del negocio y el url del sitio.
        if "<a" in line and "href" in line and "<!" not in line:
            link = BeautifulSoup(line,'lxml')
            a = link.find("a")
            Tipos.append(Tipo)
            try:
                URLs.append(a['href'])
            except:
                URLs.append("")
            Nombres.append(a['title'])
            
#Crear un dataframe con los resultados.
dataframe = {"URLs":URLs,
             "Tipo":Tipos,
             "Nombre":Nombres}

sitios = pd.DataFrame(dataframe)
sitios.head()
```

```python
new = sitios.Nombre
new.columns = ["Nombre"]
new.to_csv("Nombres.csv")
```

### 2. Clasificar en Negocios de Productos O servicios

```python
productos = ["Alimentos y Bebidas","Artesanías","Artículos Electrónicos","Boletería y Eventos","Cigarros",
            "Cine y Entretenimiento","Compras Colectivas – Cupones","Computadoras y Suministros","Domicilios en línea",
            "Cosméticos y Artículos de Belleza","Ferreterías","Floristerías","Hogar, Muebles y Electrodomésticos",
            "Joyas y Accesorios","Jugueterías","Libros y Útiles Escolares","Marketplaces",
            "Mascotas y Artículos para Mascotas","Papelería y Decoración","Repuestos y accesorios para vehículos",
            "Ropa y Accesorios","Ropa y artículos para niños","Sex Shops","Supermercados en Línea"]

def Simplificar_Tipo(tipo):
    if tipo in productos:
        return "Productos"
    else:
        return "Servicios"

sitios.Tipo = sitios.Tipo.apply(Simplificar_Tipo)
sitios
```

### 3. Revisar cada sitio para ver si usa Analytics.

```python
Usa_GAnalytics = []
#Agente de usuario para enmascarar el scrapping.
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
hdr={'User-Agent':user_agent,} 


for url in sitios.URLs:
    print("Buscando en " + url)
    try:
        req = urllib.request.Request(url,headers = hdr)
        page = urllib.request.urlopen(req).read()
        page = BeautifulSoup(page,"lxml")
        page = str(page)
        #Elementos en el HTML de la pagina que dicen si se usa analytics o no. (No es infalible)
        if "analytics.js" in page or "ga.js" in page or "analy" in page.lower():
            print("Encontro GAnalytics")
            Usa_GAnalytics.append("Si")
        else:
            print("No lo encontro")
            Usa_GAnalytics.append("No")
    except:
        print("Hubo un error en la conexion")
        Usa_GAnalytics.append("No Hubo Acceso")
sitios["Usa Analytics"] = Usa_GAnalytics
```
